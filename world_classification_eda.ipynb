### Cleaning Data

```python
import re

def clean_and_convert(value):
    try:
        # Ensure value is a string
        value = str(value)
        # Remove unwanted characters
        cleaned_value = re.sub(r'[^\d.]', '', value)
        # Convert to float if numeric, otherwise return None
        return float(cleaned_value) if cleaned_value else None
    except ValueError:
        return None

# Apply to all columns
columns_to_clean = ['GDP', 'Health Exp/Capita', 'Tourism Inbound', 'Tourism Outbound']

for col in columns_to_clean:
    df[col] = df[col].apply(clean_and_convert)
```

```python
df['Business Tax Rate']=df['Business Tax Rate'].astype(str).str.replace('%','',regex=True)
df['Business Tax Rate'] = pd.to_numeric(df['Business Tax Rate'], errors='coerce')
```

### Visualizations

```python
#box plot to check outliers
import matplotlib.pyplot as plt
plt.figure(figsize=(25,20))
nrows = 5
ncols = 5
# Get the number of numeric columns
num_cols = df.select_dtypes(include=np.number).columns
# Iterate through numeric columns and create subplots
for i, col in enumerate(num_cols):
    plt.subplot(5,5, i + 1)  # Use i + 1 as the subplot index, starting from 1
    plt.boxplot(df[col])
    plt.title(col)
plt.tight_layout()  # Adjust subplot parameters for a tight layout
plt.show()
```

Here we can see there are many outliers in the columns like  'BusinessTaxRate', 'CO2Emissions', 'Country',
       'DaystoStartBusiness', 'EaseofBusiness', 'EnergyUsage', 'GDP',
       'HealthExp%GDP', 'HealthExp/Capita', 'HourstodoTax',InfantMortalityRate','LendingInterest','MobilePhoneUsage', 'PopulationTotal', 'TourismInbound', 'TourismOutbound'.
       These can be treated by using Inter Quartile Range

```python
#Treating Outliers
def replace_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where((data[column] < lower_bound) | (data[column] > upper_bound),
                           data[column].mean(), data[column])
    return data


# Apply outlier replacement to specific columns
for col in ['GDP', 'HealthExp/Capita', 'TourismInbound', 'TourismOutbound', 'BusinessTaxRate',
            'DaystoStartBusiness', 'LifeExpectancyFemale', 'LifeExpectancyMale']:
    df = replace_outliers_iqr(df, col)
```

```python
from scipy.stats.mstats import winsorize

# Specify columns to winsorize and limits
cols_to_winsorize = ['BusinessTaxRate','CO2Emissions','EaseofBusiness','EnergyUsage','HealthExp%GDP', 'HourstodoTax',
       'InfantMortalityRate','LendingInterest','MobilePhoneUsage','Population15-64', 'Population65+', 'PopulationTotal']
limits = [0.10, 0.10]  # Winsorize 10% from both tails

# Apply winsorization
for col in cols_to_winsorize:
    df[col] = winsorize(df[col], limits=limits)
```

## Clustering

```python
#standardization
from sklearn.preprocessing import StandardScaler
numeric_cols = df.select_dtypes(include=np.number).columns
scaler = StandardScaler()
stnd_df = scaler.fit_transform(df[numeric_cols])
```

```python
#ELBOW GRAPH - To know no.of clusters to be plotted
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(stnd_df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

* The optimal number of clusters from the elbow graph is 3 .
* Because there is a sudden decrease in the graph at 3.

## 1. Hierarchical Clustering

```python
#creating dendrogram
from scipy.cluster.hierarchy import dendrogram,linkage
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,8))
dendrogram(sch.linkage(stnd_df,method='ward'))
plt.show()
```

* The above is an dendrogram plotted for hierarchical clustering
* It can be clearly understood that there are 3 clusters in the above graph since optimal number of clusters is 3.

```python
#Scatter Plot
plt.figure(figsize=(10, 7))
plt.scatter(stnd_df[:, 0], stnd_df[:, 1], c=sch.fcluster(sch.linkage(stnd_df, method='ward'), 3, criterion='maxclust'), cmap='viridis') #'viridis'
plt.title('Hierarchical Clustering Scatter Plot')
plt.xlabel('Feature 1 ')
plt.ylabel('Feature 2 ')
plt.show()
```

* If we check the scatter plot it is having 3 different types of clusters
* But they are not classified correctly We haven`t got the complete pure regions.
* So,let`s try with other type of clusters

## 2. K-Means Clustering

```python
kmeans = KMeans(n_clusters=3,random_state=42)
predict=kmeans.fit_predict(stnd_df)
```

```python
# prompt: scatter plot for kmeans

# Scatter Plot for K-Means Clustering
plt.figure(figsize=(10, 7))
plt.scatter(stnd_df[:, 0], stnd_df[:, 1], c=predict, cmap='viridis')
plt.title('K-Means Clustering Scatter Plot')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

* Here we can see from the above graph there is no change in clustering in b/w Hierarchical and K-Means.
* Since there is no change in clusters lets apply PCA to differentiate.

## PCA

```python
from sklearn.decomposition import PCA
pca = PCA()
pca_cols = pca.fit_transform(stnd_df)
```

```python
#Amount of variance
pca.explained_variance_
```

```python
#cummulative variance
var1 = np.cumsum(np.round(pca.explained_variance_ratio_,decimals=4)*100)
var1
```

```python
#variance plot to know how mmany columns should be considered
plt.plot(var1,color='blue')
plt.xlabel('Number Of Components')
plt.ylabel('Cummulative Percentage')
plt.show()
```

* By the graph we can consider 15 columns.
* It is because over 100% data, 95% of data is covered within 15 columns.

```python
data_pca = pca_cols[:,:15]
data_pca
```

```python
## Plot between PCA's
x=pca_cols[:,0]
y=pca_cols[:,1]
z=pca_cols[:,2]
plt.scatter(x,y)
plt.scatter(x,z)
plt.scatter(y,z)
plt.show()
```

```python
#joining clusters
join_clus = pd.concat([pd.DataFrame(pca_cols[:,0:3],columns=['x','y','z']), df[['Country']]], axis = 1)
join_clus
```

```python
#Scatterplot
sns.scatterplot(data = join_clus)
```

```python

## creating clusters
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(data_pca)
```

```python

plt.scatter(data_pca[y_kmeans == 0, 0], data_pca[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(data_pca[y_kmeans == 1, 0], data_pca[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(data_pca[y_kmeans == 2, 0], data_pca[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Clusters of measurements')
plt.legend()
plt.show()
```

```python
# Accuracy check
from sklearn.metrics import silhouette_score
s1_kmeans = silhouette_score(data_pca, y_kmeans)
print('Silhouette Score for K-means clustring :', s1_kmeans)
```
